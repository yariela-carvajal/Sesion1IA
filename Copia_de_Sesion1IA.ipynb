## Variables y Tipos Primitivos

En Python, una **variable** es simplemente un nombre que apunta a un valor almacenado en memoria. A diferencia de otros lenguajes (como Java o C++), en Python **no necesitas declarar el tipo** de una variable — Python lo infiere automáticamente cuando asignas un valor. Esto se llama **tipado dinámico**.

```python
x = 42        # Python sabe que x es un int
x = "hola"    # Ahora x es un str — ¡sin error!
```

Esto hace que Python sea muy flexible y rápido para prototipar, pero también requiere más cuidado: si le pasas un `str` a una función que espera un `int`, el error aparecerá en tiempo de ejecución, no al compilar.

### Tipos básicos que debes conocer

| Tipo | Ejemplo | Para qué sirve | Nota importante |
|------|---------|-----------------|-----------------|
| `int` | `42` | Números enteros | En Python no tienen límite de tamaño |
| `float` | `3.14` | Números decimales | Cuidado con la precisión: `0.1 + 0.2 ≠ 0.3` |
| `str` | `"hola"` | Texto (cadenas de caracteres) | Inmutables: no puedes cambiar un carácter suelto |
| `bool` | `True` / `False` | Valores lógicos (verdadero/falso) | Se usan en condiciones (`if`, `while`) |
| `None` | `None` | Ausencia de valor | Equivalente a `null` en otros lenguajes |

### ¿Por qué importa conocer los tipos?

Cuando le pides código a una IA, necesitas poder **especificar qué tipo de datos** esperas. No es lo mismo pedir "una función que reciba un número" que "una función que reciba una lista de diccionarios con claves `nombre` (str) y `precio` (float)". Cuanto más preciso seas con los tipos, mejor será el código generado.

````python
# Variables y tipos — lo básico
edad = 28
altura = 1.75
nombre = "María"
es_estudiante = True
dato_vacio = None
es_mayor_edad = edad >= 18

# Python infiere el tipo automáticamente
print(f"nombre: {nombre} → tipo: {type(nombre).__name__}")
print(f"edad: {edad} → tipo: {type(edad).__name__}")
print(f"altura: {altura} → tipo: {type(altura).__name__}")
print(f"es_estudiante: {es_estudiante} → tipo: {type(es_estudiante).__name__}")
print(f"dato_vacio: {dato_vacio} → tipo: {type(dato_vacio).__name__}")
````

## Operadores

Los operadores son los **bloques básicos de cualquier expresión** en Python. Aunque parezcan simples, es importante entenderlos bien porque los usaréis constantemente tanto en código propio como en código generado por IA.

### Operadores aritméticos

| Operador | Nombre | Ejemplo | Resultado |
|----------|--------|---------|-----------|
| `+` | Suma | `10 + 3` | `13` |
| `-` | Resta | `10 - 3` | `7` |
| `*` | Multiplicación | `10 * 3` | `30` |
| `/` | División (siempre float) | `10 / 3` | `3.333...` |
| `//` | División entera (trunca) | `10 // 3` | `3` |
| `%` | Módulo (resto) | `10 % 3` | `1` |
| `**` | Potencia | `10 ** 3` | `1000` |

> 💡 **Detalle importante**: la **división `/`** en Python 3 siempre devuelve un `float`, incluso si los dos números son enteros (`10 / 2` → `5.0`). Si quieres un resultado entero, usa `//`.

### Operadores de comparación

Devuelven siempre `True` o `False`. Son los que usarás en las condiciones `if`, `while`, etc.

| Operador | Significado |
|----------|-------------|
| `==` | Igual a |
| `!=` | Distinto de |
| `>`, `<` | Mayor que, menor que |
| `>=`, `<=` | Mayor o igual, menor o igual |

### Operadores lógicos

Combinan condiciones booleanas:

| Operador | Significado | Ejemplo |
|----------|-------------|---------|
| `and` | Ambas deben ser True | `x > 0 and x < 100` |
| `or` | Al menos una debe ser True | `x < 0 or x > 100` |
| `not` | Invierte el valor | `not True` → `False` |

> 🤖 **Para el prompting**: cuando pidas a la IA que genere condiciones complejas, es útil describirlas en lenguaje natural: *"filtra los productos cuyo precio sea mayor a 10 **y** cuyo stock sea mayor a 0, **o** que estén marcados como destacados"*.

````python
# Operadores en acción
a, b = 10, 3

print(f"{a} + {b} = {a + b}")
print(f"{a} ** {b} = {a ** b}")      # Potencia: 1000
print(f"{a} // {b} = {a // b}")      # División entera: 3
print(f"{a} % {b} = {a % b}")        # Resto: 1
print(f"{a} > {b} = {a > b}")        # True
print(f"{a} == {b} = {a == b}")      # False
print(f"True and False = {True and False}")  # False
print(f"not True = {not True}")              # False
````

## Strings y f-strings

Los **strings** (cadenas de texto) son uno de los tipos más utilizados en programación, y especialmente en IA, donde gran parte del trabajo implica **procesar, formatear y manipular texto** (prompts, respuestas de modelos, datos textuales, etc.).

### Formas de crear strings

```python
texto_1 = "Comillas dobles"       # La más común
texto_2 = 'Comillas simples'      # Equivalente
texto_3 = """Triple comillas       # Para texto multilínea
que ocupa varias líneas"""
```

### f-strings: la forma moderna de formatear

Introducidos en Python 3.6, los **f-strings** (formatted string literals) son la forma más legible y eficiente de insertar variables y expresiones dentro de un string. Se escriben con una `f` antes de las comillas y las variables van entre `{}`:

```python
nombre = "María"
edad = 28
print(f"Me llamo {nombre} y tengo {edad} años")
```

Dentro de los `{}` puedes poner **cualquier expresión Python válida**, no solo variables: operaciones matemáticas, llamadas a funciones, acceso a diccionarios, etc.

### Métodos útiles de strings

Los strings en Python tienen decenas de métodos incorporados. Los más importantes para trabajo con IA:

| Método | Qué hace | Ejemplo |
|--------|----------|---------|
| `.upper()` / `.lower()` | Convierte a mayúsculas/minúsculas | `"Hola".lower()` → `"hola"` |
| `.strip()` | Elimina espacios al inicio y final | `"  hola  ".strip()` → `"hola"` |
| `.split(sep)` | Divide en lista por separador | `"a,b,c".split(",")` → `["a","b","c"]` |
| `.join(lista)` | Une lista en string | `", ".join(["a","b"])` → `"a, b"` |
| `.replace(old, new)` | Reemplaza texto | `"hola mundo".replace("mundo", "Python")` |
| `.startswith()` / `.endswith()` | Comprueba inicio/final | `"test.py".endswith(".py")` → `True` |
| `.find(sub)` | Busca posición de subcadena | `"hola".find("ol")` → `1` |

> 🧠 **No necesitas memorizar** todos estos métodos. Lo importante es saber que existen para poder pedírselos a la IA: *"necesito limpiar un texto eliminando espacios y convirtiendo a minúsculas"*.

````python
# f-strings — la forma moderna de formatear texto
nombre = "Carlos"
edad = 30
nota = 8.756

# Interpolación de variables
print(f"Hola, me llamo {nombre} y tengo {edad} años")

# Expresiones dentro de f-strings
print(f"El año que viene tendré {edad + 1} años")

# Formateo de decimales
print(f"Mi nota es {nota:.2f}")  # 8.76 (2 decimales)

# Métodos útiles de strings
texto = "  Hola Mundo  "
print(f"upper: {texto.upper()}")
print(f"strip: '{texto.strip()}'")
print(f"replace: {texto.replace('Mundo', 'Python')}")
print(f"split: {'uno,dos,tres'.split(',')}")
````

## Estructuras de Datos

Las estructuras de datos son **contenedores** que nos permiten agrupar múltiples valores. Son fundamentales porque en IA trabajaréis constantemente con colecciones de datos: listas de predicciones, diccionarios con configuraciones de modelos, conjuntos de etiquetas únicas, etc.

Python tiene 4 estructuras de datos fundamentales integradas. **No necesitas memorizar todos sus métodos** — lo clave es saber **cuándo usar cada una** para poder especificarlo en tus prompts:

| Estructura | Ordenada | Mutable | Duplicados | Cuándo usarla |
|------------|----------|---------|------------|---------------|
| **Lista** `[]` | ✅ | ✅ | ✅ | Colección general de elementos (la más versátil) |
| **Diccionario** `{}` | ✅* | ✅ | Claves únicas | Datos nombrados: clave → valor (como un JSON) |
| **Tupla** `()` | ✅ | ❌ | ✅ | Datos que no deben cambiar (coordenadas, constantes) |
| **Set** `{}` | ❌ | ✅ | ❌ | Eliminar duplicados, buscar pertenencia rápidamente |

### ¿Mutable vs Inmutable?

- **Mutable** significa que puedes modificar la estructura después de crearla (añadir, eliminar, cambiar elementos).
- **Inmutable** significa que una vez creada, no puedes modificarla. Si necesitas una "versión diferente", tienes que crear una nueva.

Las tuplas son inmutables, lo que las hace útiles para datos que representan algo fijo (como coordenadas geográficas) y como claves de diccionario (las listas no pueden ser claves).

### Listas vs Diccionarios: la decisión más común

En la práctica, las dos estructuras que más usaréis con diferencia son las **listas** y los **diccionarios**:

- Usa una **lista** cuando tengas una **secuencia** de elementos del mismo tipo: `[1, 2, 3]`, `["manzana", "pera", "uva"]`.
- Usa un **diccionario** cuando necesites asociar **nombres a valores**: `{"nombre": "Ana", "edad": 22}` — es la estructura más parecida a un JSON, y la usaréis constantemente con APIs, modelos de IA y configuraciones.

### Slicing (rebanado) en listas

Una de las características más potentes de Python es el **slicing**, que permite extraer sub-secciones de listas y strings:

```python
datos = [10, 20, 30, 40, 50]
datos[1:3]    # [20, 30]     — desde posición 1 hasta 3 (sin incluir)
datos[:2]     # [10, 20]     — desde el inicio hasta 2
datos[-2:]    # [40, 50]     — los últimos 2 elementos
datos[::2]    # [10, 30, 50] — de 2 en 2
```

> 🤖 **Para el prompting**: cuando pidas código a la IA, especifica siempre qué estructura de datos quieres: *"devuelve una lista de diccionarios"*, *"agrupa los resultados en un diccionario donde las claves sean las categorías"*, etc.

## Control de Flujo

El **control de flujo** determina el orden en que se ejecutan las instrucciones de tu programa. Sin control de flujo, el código se ejecutaría de arriba a abajo sin tomar decisiones ni repetir acciones — lo cual sería bastante limitado.

Python ofrece tres mecanismos principales:

- **Condicionales** (`if` / `elif` / `else`): el programa toma **decisiones** diferentes según se cumplan o no ciertas condiciones.
- **Bucles `for`**: repiten un bloque de código **una vez por cada elemento** de una secuencia (lista, rango, string...).
- **Bucles `while`**: repiten un bloque de código **mientras** se cumpla una condición.

### Indentación: cómo Python sabe qué código va dentro de qué

A diferencia de otros lenguajes que usan `{}` para delimitar bloques, **Python usa la indentación** (espacios al inicio de la línea). Todo el código que esté indentado bajo un `if`, `for` o `def` pertenece a ese bloque. Esto fuerza un estilo limpio, pero también significa que un espacio mal puesto puede causar errores o cambiar la lógica del programa.

```python
if condicion:
    # Este código se ejecuta si condicion es True
    hacer_algo()
    # Sigue dentro del if
# Este código ya está fuera del if (sin indentación)
```

### List comprehensions: la forma "Pythónica" de crear listas

Una de las características más elegantes de Python es la **list comprehension**, que permite crear listas en una sola línea a partir de un bucle:

```python
# Forma clásica (4 líneas)
cuadrados = []
for n in range(10):
    cuadrados.append(n ** 2)

# List comprehension (1 línea)
cuadrados = [n ** 2 for n in range(10)]

# Con filtro
pares = [n for n in range(20) if n % 2 == 0]
```

Las list comprehensions son muy comunes en código Python y en respuestas de IA — si las veis en código generado, ahora sabéis qué hacen.

## Funciones

Las funciones son la herramienta clave para **organizar y reutilizar código**. Una función encapsula un bloque de lógica con un nombre, recibe datos de entrada (parámetros) y produce un resultado (valor de retorno).

### ¿Por qué son tan importantes?

1. **Reutilización**: escribes la lógica una vez y la usas cuantas veces quieras con distintos datos.
2. **Legibilidad**: `calcular_imc(70, 1.75)` es mucho más claro que un bloque de 10 líneas de cálculos.
3. **Modularidad**: un programa complejo se descompone en funciones pequeñas y manejables.
4. **Testing**: es mucho más fácil verificar que una función funciona correctamente que probar un programa entero.

### Anatomía de una función Python

```python
def nombre_funcion(parametro1, parametro2="valor_default"):
    """Docstring: breve descripción de qué hace la función."""
    # Cuerpo de la función
    resultado = parametro1 + parametro2
    return resultado  # Devuelve el resultado
```

- `def` indica que estás definiendo una función.
- Los **parámetros** entre paréntesis son los datos de entrada. Pueden tener valores por defecto.
- El `return` devuelve el resultado. Si no hay `return`, la función devuelve `None`.
- El **docstring** (entre triple comillas) documenta la función — es lo que aparece al hacer `help(nombre_funcion)`.

### Parámetros especiales: `*args` y `**kwargs`

A veces necesitas funciones que acepten un **número variable** de argumentos:

- `*args`: recibe argumentos posicionales como una tupla → `def f(*args)` → `f(1, 2, 3)` → `args = (1, 2, 3)`
- `**kwargs`: recibe argumentos con nombre como un diccionario → `def f(**kwargs)` → `f(a=1, b=2)` → `kwargs = {"a": 1, "b": 2}`

Veréis esto mucho en librerías de IA (Keras, HuggingFace, etc.) donde las funciones aceptan docenas de parámetros opcionales.

> 🤖 **Consejo de prompting**: cuando pidas una función a la IA, especifica: el **nombre**, los **parámetros** (con tipos de ser posible), qué debe **devolver**, y si necesitas **validación de errores** o **docstring**.

````python
# Función básica
def saludar(nombre):
    return f"¡Hola, {nombre}!"

print(saludar("María"))

# Función con parámetros por defecto
def calcular_precio(precio, descuento=0, iva=0.21):
    precio_descuento = precio * (1 - descuento)
    precio_final = precio_descuento * (1 + iva)
    return round(precio_final, 2)

print(f"Sin descuento: {calcular_precio(100)}€")
print(f"Con 10% dto: {calcular_precio(100, descuento=0.10)}€")

# Función con *args y **kwargs
def resumen(*notas, nombre="Anónimo"):
    media = sum(notas) / len(notas) if notas else 0
    return f"{nombre}: media = {media:.1f} ({len(notas)} notas)"

print(resumen(8, 7, 9, nombre="Carlos"))
print(resumen(6, 5, 7, 8))
````

# Bloque 3 — Introducción al Prompting para Desarrollo

> 🎯 El prompting no es solo una habilidad para hablar con ChatGPT. Es la forma en que **cualquier desarrollador moderno** interactúa con herramientas de IA para escribir código.

### ¿Qué es el prompting?

El **prompting** es el arte de formular instrucciones precisas a un modelo de lenguaje (LLM) para obtener el resultado deseado. En el contexto del desarrollo de software, esto significa **escribir descripciones en lenguaje natural** que la IA traduce a código.

Pensad en ello como la diferencia entre decirle a un taxista "llévame al centro" (vago, muchas interpretaciones posibles) vs "llévame a la Gran Vía 42, esquina con calle Hortaleza" (preciso, resultado predecible).

### ¿Por qué es una habilidad crítica?

Según estudios recientes del sector (GitHub, Stack Overflow Developer Survey 2024), **más del 70% de los desarrolladores** ya usan herramientas de IA para escribir código. La productividad no depende de la herramienta en sí, sino de la **calidad de los prompts** que el desarrollador escribe.

Un buen prompt puede generar código funcional, documentado y con manejo de errores en segundos. Un mal prompt genera código genérico que tendrás que reescribir de todos modos.

## Anatomía de un Buen Prompt para Código

Un buen prompt de desarrollo es como una **especificación técnica** — cuanto más completa, mejor será el resultado. Estos son los componentes que debe incluir:

```
┌─────────────────────────────────────────────┐
│  1. INSTRUCCIÓN CLARA                       │
│     "Crea una función que..."               │
│                                             │
│  2. CONTEXTO                                │
│     "...para un sistema de gestión de       │
│      inventario..."                         │
│                                             │
│  3. ESPECIFICACIONES                        │
│     "...que reciba una lista de productos   │
│      (nombre, precio, stock)..."            │
│                                             │
│  4. RESTRICCIONES / FORMATO                 │
│     "...devuelva un diccionario, usa type   │
│      hints, incluye manejo de errores."     │
│                                             │
│  5. EJEMPLOS (opcional - Few-shot)          │
│     "Ejemplo: entrada=[...] → salida={...}" │
└─────────────────────────────────────────────┘
```

### Desglose de cada componente:

**1. Instrucción clara**: el verbo que indica qué acción debe realizar la IA. "Crea", "Modifica", "Explica", "Depura", "Optimiza", "Convierte"... Evita ser ambiguo — "Haz algo con datos" no es una instrucción.

**2. Contexto**: explica el **dominio** o **proyecto** en el que se enmarca el código. Esto ayuda a la IA a elegir nombres de variables, estructuras y patrones adecuados. No es lo mismo código para un sistema médico que para un videojuego.

**3. Especificaciones**: los **tipos de datos** de entrada y salida, las **reglas de negocio** que debe cumplir, y el **comportamiento esperado**. Esta es la parte más importante y la que más gente omite.

**4. Restricciones / Formato**: indicaciones sobre el estilo del código, librerías permitidas (o prohibidas), formato de salida, idioma de las variables, etc.

**5. Ejemplos (Few-shot)**: proporcionar uno o dos ejemplos concretos de entrada → salida esperada. Esto es especialmente útil para transformaciones de datos o lógica compleja donde el lenguaje natural puede ser ambiguo.

> 🔑 **Regla de oro**: si tu prompt tiene menos de 2 líneas, probablemente le falta información. Un buen prompt suele tener entre 3 y 8 líneas.

````python
def calcular_imc(peso, altura):
    """
    Calcula el Índice de Masa Corporal (IMC) y determina la categoría.

    Args:
        peso (float): Peso en kg.
        altura (float): Altura en metros.

    Returns:
        dict: Un diccionario con el valor del IMC y la categoría.
    """
    if peso <= 0 or altura <= 0:
        return "Error: El peso y la altura deben ser valores positivos."

    imc = round(peso / (altura ** 2), 1)

    if imc < 18.5:
        categoria = "Bajo peso"
    elif 18.5 <= imc < 25:
        categoria = "Normal"
    elif 25 <= imc < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"

    return {"imc": imc, "categoria": categoria}

# Ejemplo de prueba
resultado_imc = calcular_imc(75, 1.80)
print(f"Resultado del análisis: {resultado_imc}")
````

````python
def calcular_imc(peso: float, altura: float) -> dict:
    """
    Calcula el Índice de Masa Corporal (IMC) y determina la categoría de salud.

    Args:
        peso (float): Peso del individuo en kilogramos.
        altura (float): Altura del individuo en metros.

    Returns:
        dict: Diccionario con el 'imc' redondeado y la 'categoria' correspondiente.

    Raises:
        ValueError: Si el peso o la altura no son positivos, o si la altura supera los 3 metros.
    """
    # Validaciones
    if peso <= 0 or altura <= 0:
        raise ValueError("El peso y la altura deben ser valores positivos.")
    if altura > 3:
        raise ValueError("La altura no puede ser superior a 3 metros.")

    # Cálculo del IMC
    valor_imc = round(peso / (altura ** 2), 1)

    # Determinación de categoría
    if valor_imc < 18.5:
        categoria = "Bajo peso"
    elif 18.5 <= valor_imc < 25:
        categoria = "Normal"
    elif 25 <= valor_imc < 30:
        categoria = "Sobrepeso"
    else:
        categoria = "Obesidad"

    return {
        "imc": valor_imc,
        "categoria": categoria
    }

# Ejemplos de prueba
try:
    print(f"Caso normal: {calcular_imc(70.5, 1.75)}")
    print(f"Caso sobrepeso: {calcular_imc(90.0, 1.80)}")
    # print(calcular_imc(-5, 1.70)) # Esto lanzaría ValueError
except ValueError as e:
    print(f"Error de validación: {e}")
````

## Comparativa: Prompt Malo vs Prompt Bueno

La diferencia entre un prompt mediocre y uno excelente no es cuestión de estilo — es la diferencia entre recibir código genérico que no te sirve y recibir exactamente lo que necesitas.

### ❌ Prompt malo
> "Haz una función de IMC"

**¿Qué recibirás?** Una función básica que calcula `peso / altura²`. Sin validación, sin categorías, sin type hints, probablemente con nombres de variables genéricos. Tendrás que reescribirla tú mismo.

### ✅ Prompt bueno
> "Crea una función en Python llamada `calcular_imc` que:
> - Reciba peso (en kg) y altura (en metros) como float.
> - Devuelva un diccionario con el valor del IMC redondeado a 1 decimal y la categoría (Bajo peso, Normal, Sobrepeso, Obesidad).
> - Incluya validación: peso y altura deben ser positivos, la altura no puede ser mayor a 3 metros.
> - Si los datos no son válidos, lance un ValueError con mensaje descriptivo.
> - Usa type hints y añade un docstring."

**¿Qué recibirás?** Exactamente lo que has pedido — una función robusta, documentada y con manejo de errores. Código que podrías usar en producción directamente.

### ¿Por qué funciona mejor?

El segundo prompt es mejor porque:
1. **Nombra la función** → evita que la IA invente un nombre genérico.
2. **Especifica tipos** → `float`, no "números".
3. **Define la salida** → diccionario con claves concretas, no "algo".
4. **Incluye reglas de negocio** → los rangos de las categorías.
5. **Pide validación** → manejo de errores con tipo de excepción específico.
6. **Indica estilo** → type hints y docstring.

Cada una de estas decisiones la tendrías que tomar tú de todos modos si programaras la función a mano. La diferencia es que ahora se las comunicas a la IA y ella hace el trabajo pesado.

## Técnicas Básicas de Prompting

Existen varias técnicas reconocidas en la literatura de prompting que podéis aplicar desde hoy. En la Sesión 2 profundizaremos en técnicas avanzadas como Chain of Thought y Role Prompting, pero estas son las fundamentales:

| Técnica | Qué es | Ejemplo | Cuándo usarla |
|---------|--------|---------|---------------|
| **Instrucción directa** | Decir exactamente qué quieres | *"Crea una función que..."* | Siempre — es la base |
| **Contexto / Rol** | Dar contexto al modelo | *"Eres un desarrollador Python senior..."* | Cuando necesitas un estilo o nivel específico |
| **Few-shot** | Dar ejemplos de entrada/salida | *"Ejemplo: [1,2,3] → 6"* | Cuando la lógica es ambigua en lenguaje natural |
| **Restricciones** | Limitar el resultado | *"No uses librerías externas"* | Para controlar la solución |
| **Paso a paso** | Pedir que razone antes de codificar | *"Primero explica el algoritmo, luego codifica"* | Para problemas complejos |
| **Iteración** | Refinar el resultado en múltiples turnos | *"Ahora añade manejo de errores"* | Siempre — es el flujo natural |

### Profundizando en cada técnica:

**Instrucción directa**: la más simple y universal. Consiste en pedir claramente lo que necesitas. El truco está en ser **específico**: en vez de "ordena una lista", di "ordena una lista de diccionarios por el campo 'fecha' de más reciente a más antigua".

**Contexto / Rol**: al asignar un rol al modelo ("Eres un experto en seguridad informática", "Actúa como un profesor de Python explicando a principiantes"), cambias el **tono**, la **terminología** y el **nivel de detalle** de la respuesta.

**Few-shot**: una de las técnicas más poderosas. Cuando es difícil describir una transformación con palabras, dale 2-3 ejemplos y la IA inferirá el patrón. En la Sesión 3 veréis cómo esto se implementa programáticamente con LangChain.

**Iteración**: este es el flujo de trabajo más realista. Casi nunca obtendrás el código perfecto al primer intento. El proceso es: pedir → revisar → refinar → probar → refinar más. Cada turno de conversación con la IA os acerca al resultado deseado.

> 💡 **En la práctica, usaréis varias técnicas a la vez**: contexto + instrucción directa + restricciones es la combinación más habitual.

## Patrones Anti-Prompt — Qué NO hacer

Tan importante como saber escribir buenos prompts es reconocer los **errores más comunes** que reducen drásticamente la calidad del código generado:

| ❌ Anti-patrón | Por qué falla | ✅ Mejor alternativa |
|----------------|---------------|---------------------|
| "Haz algo con datos" | Demasiado vago — la IA tiene que inventar todo | "Crea una función que reciba una lista de diccionarios con campos nombre, edad, nota y devuelva..." |
| Pedir todo de una vez | El modelo se confunde con instrucciones largas y complejas | Ir paso a paso e iterar: primero la lógica básica, luego validaciones, luego documentación |
| No especificar formato | La IA elige un formato arbitrario que quizás no te sirve | "Devuelve un JSON / diccionario / DataFrame / lista de tuplas..." |
| Copiar sin leer | Puedes integrar errores, vulnerabilidades o código ineficiente | Leer, entender, probar con diferentes inputs y luego integrar |
| No dar contexto | El código generado será genérico y desconectado de tu proyecto | "Para una API REST con FastAPI..." / "Para un análisis en Colab..." |
| No probar los edge cases | El código funciona con el ejemplo pero falla con datos reales | "¿Qué pasa si la lista está vacía?", "¿Y si hay valores None?" |

### El "Síndrome del Copy-Paste"

Uno de los mayores riesgos del desarrollo asistido por IA es caer en lo que llamaremos el **"Síndrome del Copy-Paste"**: generar código, copiarlo, ejecutarlo, ver que da un resultado, y asumir que es correcto. Este patrón es peligroso porque:

1. El código puede funcionar con tus datos de prueba pero fallar con datos reales.
2. Puede tener errores lógicos sutiles que no se manifiestan inmediatamente.
3. Puede usar malas prácticas (variables globales, código duplicado, vulnerabilidades de seguridad).

**La solución**: desarrollar el hábito de **leer y entender** cada línea del código generado. Si hay algo que no entiendes, **pide a la IA que te lo explique** — eso es aprender.

## ¿Por qué Git y GitHub?

### Git — Control de versiones

**Git** es un sistema de control de versiones distribuido. En términos simples, es una herramienta que:

- 📝 Guarda el **historial completo** de cambios de tu código — cada modificación queda registrada con fecha, autor y descripción.
- 🔀 Permite que **varias personas** trabajen en el mismo proyecto sin pisarse entre sí, gracias al sistema de **ramas** (branches).
- ↩️ Te permite **volver atrás** si algo sale mal — puedes recuperar cualquier versión anterior de tu código.
- 🧪 Facilita la **experimentación**: puedes crear una rama para probar algo nuevo sin afectar al código principal.

Pensad en Git como un sistema de "guardado inteligente" que, a diferencia del Ctrl+Z, funciona entre sesiones, entre ordenadores y entre personas.

### GitHub — La plataforma

**GitHub** es la plataforma web más popular para alojar repositorios Git. Es mucho más que almacenamiento — es el **ecosistema central** del desarrollo de software y la IA:

- **Código abierto**: todos los proyectos importantes de IA están en GitHub: HuggingFace Transformers, LangChain, PyTorch, TensorFlow, Keras, Gradio, Streamlit...
- **Colaboración**: sistema de Issues (reportar problemas), Pull Requests (proponer cambios), Code Review (revisar código de otros).
- **Portfolio profesional**: tu perfil de GitHub es tu CV técnico. Los reclutadores lo miran antes de tu LinkedIn.
- **CI/CD**: automatización de tests, despliegue y publicación de paquetes (lo veréis en MLOps).
- **GitHub Copilot**: el agente de IA para desarrollo que veremos en la Sesión 2 está profundamente integrado con GitHub.

## Conceptos Clave de Git

Para trabajar con Git, hay que entender los **tres espacios** donde vive tu código y cómo se mueve entre ellos:

```
┌──────────────────────────────────────────────────────────┐
│                     FLUJO BÁSICO DE GIT                  │
│                                                          │
│   Working Dir   →    Staging Area    →    Repository     │
│   (tus archivos)     (git add)            (git commit)   │
│                                                          │
│   "Estoy editando"  "Esto está listo"   "Guardado en    │
│                                          el historial"   │
│                                                          │
│   Local         ────────────────────→    GitHub (remoto) │
│                       git push                           │
│                                                          │
│   GitHub        ────────────────────→    Local           │
│                       git pull                           │
└──────────────────────────────────────────────────────────┘
```

### Analogía: el flujo de Git es como enviar un paquete

1. **Working Directory** = tu escritorio donde estás trabajando en el proyecto.
2. **`git add`** = metes en una caja los archivos que quieres enviar (staging area).
3. **`git commit`** = sellas la caja con una nota descriptiva y la guardas en el almacén local.
4. **`git push`** = envías la caja al almacén remoto (GitHub) para que otros puedan acceder.
5. **`git pull`** = traes a tu escritorio las cajas que otros han enviado al almacén remoto.

### Comandos esenciales

| Comando | Qué hace | Cuándo lo usas |
|---------|----------|----------------|
| `git init` | Inicia un repositorio Git en la carpeta actual | Solo una vez, al crear el proyecto |
| `git clone <url>` | Descarga un repositorio completo de GitHub | Para empezar a trabajar con un proyecto existente |
| `git status` | Muestra qué archivos han cambiado | Constantemente — es tu panel de control |
| `git add .` | Añade todos los cambios al staging | Antes de cada commit |
| `git commit -m "mensaje"` | Guarda los cambios con un mensaje descriptivo | Cada vez que completas una unidad de trabajo |
| `git push` | Sube los commits locales a GitHub | Cuando quieres publicar tus cambios |
| `git pull` | Descarga los cambios de GitHub a tu máquina | Al empezar a trabajar, para estar al día |

### ¿Cómo escribir buenos mensajes de commit?

Los mensajes de commit son la **documentación** del historial de tu proyecto. Un buen mensaje describe **qué** cambiaste y **por qué**:

- ❌ `"cambios"`, `"fix"`, `"asdfgh"` → no dicen nada
- ✅ `"Añade validación de inputs en calcular_imc"` → claro y descriptivo
- ✅ `"Corrige error de división por cero en media_notas"` → explica el problema resuelto
